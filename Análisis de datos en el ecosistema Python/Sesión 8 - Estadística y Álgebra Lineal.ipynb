# Estadística

Podríamos decir que la **Estadística** es la ciencia que se encarga del estudio de los fenómenos aleatorios, y se divide en dos grandes áreas:

- **Estadística descriptiva**, que tiene como objetivo el tratamiento de los datos con la finalidad de poner de manifiesto sus características más relevantes y sintetizarlas en unos pocos valores o mediante representaciones.
Algunos parámetros estadísticos son la media, la desviación típica o un histograma.

- **Estadística inferencial** que se centra en el análisis con el fin de obtener conclusiones que, con un margen de confianza conocido, sean extrapolables a todo el conjunto de datos que posea una determinada característica en estudio.
Algunos ejemplos de inferencia estadística son  las pruebas de hipótesis o estimación paramétrica.


La estadística inferencial se divide, a su vez, en dos subramas:
- **Estadística paramétrica**, que  comprende los procedimientos estadísticos y de decisión que están basados en distribuciones conocidas. Estas son determinadas usando un número finito de parámetros, y puede resolver 3 tipos de problemas:
 - **Estimación puntual**: En la que pretendemos darle un valor al parámetro a estimar.
 - **Estimación por intervalos** (buscamos un intervalo de confianza).
 - **Contraste de hipótesis** donde buscamos contrastar información acerca del parámetro.


- **Estadística no paramétrica** estudia las pruebas y modelos estadísticos cuya distribución subyacente no se ajusta a los llamados criterios paramétricos. Su distribución no puede ser definida a priori, pues son los datos observados los que la determinan. La utilización de estos métodos se hace recomendable cuando no se puede asumir que los datos se ajusten a una distribución conocida, cuando el nivel de medida empleado no sea, como mínimo, de intervalo

## 1. Conceptos básicos

La **media aritmética** es un promedio estándar que a menudo se denomina promedio.

La **media aritmética ponderada** de un conjunto de datos es un promedio en el que cada uno de ellos tiene una importancia relativa (o peso) respecto de los demás datos.

La **media geométrica**  es la raíz n-ésima del producto de todos los números. Es un promedio muy útil en conjuntos de números que son interpretados en orden de su producto, no de su suma (tal y como ocurre con la media aritmética). Por ejemplo, las velocidades de crecimiento. Calcular medias sobre porcentajes. Un fondo de inversión tiene una rentabilidad anual del 15%, 35% y 20%

Media geométrica = ∛(1.15∗1,35∗1,20).

La **mediana** es el valor de la variable estadística que deja igual número de observaciones por debajo y por encima de él al ordenar los datos de menor a mayor..
- Si la cantidad de valores es impar, la mediana es el valor central
- Si la cantidad de valores es par, la mediana es el valor promedio de los dos datos centrales.

La **moda** es el valor con mayor frecuencia de un conjunto de datos. Puede haber más de una moda.


El **Rango** es el intervalo entre el valor máximo y el valor mínimo. Permite obtener una idea de la dispersión de los datos.


Los **cuantiles** suelen usarse por grupos que dividen la distribución en partes iguales; entendidas estas como intervalos que comprenden la misma proporción de valores. Los más usados son:

- Los cuartiles, que dividen al conjunto de datos ordenados en cuatro partes (corresponden a los cuantiles 0,25; 0,50 y 0,75).

- El rango intercuartílico o rango intercuartil es la diferencia entre el tercer y el primer cuartil de una distribución. Es una medida de la dispersión estadística. 

- Los quintiles, que al conjunto de datos ordenados en cinco partes (corresponden a los cuantiles 0,20; 0,40; 0,60 y 0,80).

- Los deciles, que dividen al conjunto de datos ordenados en diez partes.

- Los percentiles, que al conjunto de datos ordenados en cien partes.


Cuando la mayoría de los valores de la variable se distribuyen muy cerca de los valores centrales, la serie es **concentrada**, y en caso contrario, es **dispersa**.

La **varianza** es una medida de dispersión que se utiliza para representar la variabilidad de un conjunto de datos respecto de la media aritmética de los mismos.

La **desviación** típica es la desviación media de una variable respecto de su media aritmética.






- La desviación estándar es la raíz de la varianza.

- Para calcular la varianza estamos los residuos al cuadrado, ya que de otra manera el resultado del sumatorio sería 0. Esto hace que las unidades de la varianza no sean las mismas que las de los datos originales, por eso usamos la desviación estándar.

- Si tenemos un conjunto de datos sobre las alturas de los madrileños, la varianza nos daría las variaciones sobre la media en metros cuadrados, lo que no siempre tiene sentido. La desviación estándar tiene las mismas unidades que la media.


## Probabilidad

La **probabilidad** de un suceso se define por la regla de Laplace:

P(A) = Número de casos favorables / Número de casos posibles

- La probabilidad de un suceso esta comprendida entre 0 y 1
- 1 es la probabilidad del suceso seguro y 0 es la probabilidad del suceso imposible




## Distribución Normal

Podemos definir una **distribución de probabilidad** como una lista que nos proporciona todos los resultados de los valores que pueden presentarse en un acontecimiento, junto con la probabilidad de ocurrencia asociada a cada uno de estos valores.





- Es simétrica respecto a su media μ
- La ordenada es máxima en x = μ
- La moda y la mediana son iguales a la media.
- Tiene dos puntos de inflexión en μ+ σ y μ - σ
- El área entre la función y el eje horizontal es 1 (100% población).
- Es asintótica al eje de abscisas







### Inferencia Estadística

El objetivo de la inferencia estadística es extraer conclusiones a partir de los datos muestrales para poder inferir sobre la población, y se divide en dos partes:

- Estimación, cuyo objetivo es obtener el valor de los parámetros poblacionales, de forma más o menos aproximada, a través de dos métodos que tienen objetivos distintos.
    - Consta a su vez, de la estimación puntual y de la estimación por intervalos de confianza.

- Y del contraste o test de hipótesis.


La **Población** es el conjunto de todos los individuos que constituyen el objeto de un determinado estudio, y de los cuales se quieren obtener ciertas conclusiones. 

Una **Muestra** es un subconjunto de individuos u objetos de una población, y deber ser representativa de esa población.

El **Muestreo** es el proceso de seleccionar un conjunto de individuos representativos de una población con el fin de estudiarlos y poder caracterizar el total de la población.











### Contraste de hipótesis

Un **contraste o test de hipótesis** es una regla de decisión, que nos permite decidir si determinadas hipótesis establecidas sobre el valor de algún parámetro o de una distribución son ciertas o no, de tal forma que, aceptamos o rechazamos el cumplimiento de dichas hipótesis.
- **H0 o Hipótesis nula:** Premisa que queremos comprobar y suponemos cierta (la que vamos a aceptar o rechazar)
- **H1 Hipótesis alternativa:** La  aceptaremos como válido cuando 𝐻_0 no sea cierta


El **estadístico de contraste** es el estadístico usado para decidir qué hipótesis aceptar o rechazar.

1. Los pasos para realizar el contraste de hipotesis serían:
2. Identificar el parámetro que vamos a estudiar: μ, σ, σ^2
3. Especificar las hipótesis nula y alternativa
4. Fijar un valor para el nivel de significación
5. Obtener el valor del estadístico para la muestra elegida
6. Determinar la región de aceptación y la región de rechazo
7. Decidir si rechazamos o no la hipótesis nula
8. Interpretar los resultados obtenidos.











# 2. Álgebra lineal

Se puede operar con matrices, resolver sistemas de ecuaciones...:
    
| Función | Descripción |
|:---------|:-----|
| *dot(mat1, mat2)* | Devuelve el producto escalar entre dos arrays. Si son matrices 2D, es equivalente a la multiplicación de ambas |
| *matmul(mat1, mat2)* | Devuelve el producto entre dos matrices |
| *trace(mat1, mat2)* | Suma de las diagonales de ambas matrices |
| *det(mat)* | Devuelve el determinante de la matriz |
| *eig(mat)* | Computa los autovalores y autovectores de la matriz cuadrada *mat* |
| *inv(mat)* | Devuelve la inversa de la matriz |
| *qr(mat)* | Computa la factorización QR de *mat* |
| *solve(A, b)* | Resuelve el sistema lineal de ecuaciones *Ax = b* para *b*, cuando *A* es una matriz cuadrada |

La **matriz traspuesta de una matriz A** se obtine intercambiando filas por columnas.

La **matriz inversa de una matriz A** es aquella que multiplicada por A da como resultado la matriz identidad.

El **determinante de una matriz** es un valor representativo de la misma que nos ayudará a averiguar si sus filas o columnas son linealmente independientes entre sí, o para determinar si una matriz
tiene inversa.

El **rango** de una matriz es el número de filas o columnas linealmente independientes.

Una matriz es **diagonal** si todos los valores son nulos salvo los de la diagonal.

Una matriz A es **diagonalizable** si y solo si se existe una matriz P tal que: P^-1*A*P = D, siendo D una matriz diagonal

```python
import numpy as np
import numpy.linalg as linalg

# resolver sistema de ecuaciones
# 6x + 4y = 24
# x - 3y = 2
# Ax = b --> resolver para x

a = np.array([ [3,2],[1,-3] ])
b  = np.array([12,2])
x,y = linalg.solve(a,b)
print('x=' + str(x) + ', y=' + str(y))
```

```python
A = np.arange(12).reshape(4,3)
B = np.arange(9).reshape(3,3)

print(np.matmul(A,B))
print(np.dot(A, B))
print(A @ B)
```

Siendo v el vector propio y λ el valor propio.

Los valores propios se calculan resolviendo la ecuación resultante de:

Av –λ* v =0 --> (A-*Id)*v=0 --> det(A- λ a*Id)=0
