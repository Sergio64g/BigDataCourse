**Neurona**

La función principal de las neuronas es la transmisión de impulsos nerviosos. Estos viajan por toda la neurona comenzando por las dendritas hasta llegar a las terminaciones del axón, donde pasan a otra neurona por medio de la conexión sináptica. Las neuronas son procesadores de información sencillos. Poseen un canal de entrada (dendritas), un órgano de cómputo (soma) y un canal de salida (axón).





Los ordenadores están compuestos por un único procesador, que suele ser muy potente. El cerebro, en cambio, no tiene un único procesador potente, si no que tiene millones de ellos, aunque son muy elementales, las neuronas. Éstas son mucho más simples, lentas y menos fiables que un ordenador convencional, pero resuelven ciertos problemas muy sencillos de forma mucho más eficaz, como distinguir un perro de una gato… La idea es que los ANS intentan construir sistemas que copien la estructura de las redes neuronales biológicas para poder realizar de forma eficaz la misma funcionalidad,

Los 3 conceptos clave que intentan emular las NN son:

- *Paralelismo de cálculo:* operar en paralelo en diferentes sectores de la imagen.
- *Memoria distribuida:* si una sinapsis resulta dañada, se perdería una parte muy pequeña de la información, y como los sistemas neuronales biológicos, son redundantes, ie: tolerante a fallos.
- *Adaptabilidad al entorno:* que es lo que llamamos generalización. Las NN se adaptan al entorno modificando su sinapsis y aprendiendo de la experiencia.

------------------------------------------------

Una **Red Neuronal Artificial (RNA)** es un modelo matemático inspirado en el comportamiento biológico de las neuronas y en cómo se organizan formando la estructura del cerebro. Las neuronas se unen para formar una red neuronal y son capaces de aprender, es la unidad básica de procesamiento.

Las **capas de una red neuronal** son:

- *Intup Layer:* las neuronas que asimilan los datos de entrada, como por ejemplo imagen o una tabla de datos.
- *Hidden Layer:* realiza el procesamiento de información y hacen los cálculos intermedios. Cada más neuronas en esta capa haya, más complejos son los cálculos que se efectúan.
- *Output Layer:* es el último eslabón de la cadena, y es la red que toma la decisión o realiza alguna conclusión aportando datos de salida.

El **output** de una neurona es input*weight + bias.

Se tiene bias por cada neurona y weight por cada conexión.



------------------------------------------------

El **perceptrón simple**  es una red neuronal artificial unidireccional, compuesto por dos capas de neuronas, una de entrada y otra de salida. Resuelve problemas lineales.



El **perceptrón multicapa** es una red neuronal artificial (RNA) formada por múltiples capas, de tal manera que tiene capacidad para resolver problemas que no son linealmente separables. El perceptrón multicapa puede estar totalmente o localmente conectado. En el primer caso cada salida de una neurona de la capa "i" es entrada de todas las neuronas de la capa "i+1", mientras que en el segundo cada neurona de la capa "i" es entrada de una serie de neuronas (región) de la capa "i+1".


--------------------------------

Un algoritmo encuentra los bias y los weights con ayuda de *forward / backward propagation*.

**Forward Propagation** hace referencia a cómo la información de entrada se va propagando por la red neuronal desde la capa de entrada a la capa de salida. Cada capa oculta acepta los datos de entrada, los procesa según la función de activación y los pasa a la capa sucesiva. Para generar la salida los datos deben propagarse hacia adelante, para evitar formar círculos.

**Backward Propagation** hace referencia al método usado para ajustar los pesos y minimizar la función de coste.

------------------------------------------------



**Hiperpárametros utilizados en el entrenamiento de NN**

- **Epoch:** cada época es una pasada de todos las training samples
- **Batch size:** agrupación de training samples considerados para hacer una actualización de los parámetros de la red. Puede variar entre 1 y el tamaño del training set.
 - Batch size = 1: se realizarán tantas actualizaciones de pesos como instancias en el training set, lo que provocará una convervencia más lenta. Además la actualización de pesos será más sensible a ouliers.
 - Bath size = Training set size: se realizará una única actualización de pesos retropropagando el error medio de todas las mustras, la épco es más rápida (solo hay un paso por época), pero tiene que haber más épocas para garantizar la convergencia.
 - 1 < Batch size < Training set size: solución más habitual, suele escogerse un número potencia de 2. Los pasos por época serán el Training_dataset_size / Batch_size)
- **Learning rate:** magnitud del cambio en cada actualización de parámetros
- **Función de activación:** introduce las no linealidades en el esquema de red neuronal
- **Función de pérdida:** error entre lo esperado (etiqueta, ground truth) y lo predicho en la época actual
- **Topología de la red neuronal:** número de capas ocultas y unidades/neuronas por capa


La **función de activación** es la que va a permitir introducir no linealidades en el modelo de tal forma que las ANNs puedan aprender decision boundaries que no son lineales.  



 - https://en.wikipedia.org/wiki/Activation_function

El **optimizador** es el método que se utiliza para optimizar los hiperparámetros (bias & weights) y minimizar la función de coste.

- https://keras.io/optimizers/


La **regularización** se refiere a un conjunto de técnicas que tienen por objetivo reducir overfitting. Algunas de las más utilizadas son:

- **Weight Regularization:**  coste proporcional al valor absoluto (L1) o proporcional a la raiz cuadrada del valor (L2).
- **Dropout:** consiste en desactivar neuronas aleatoriamente.
- **Data augmentation:** consiste en crear datos sintéticos y agregarlos al conjunto de entrenamiento.
- **Batcn normalization:** resta la media y divide entre la desviación típica de los datos del batch -> Estandarización de features. Se suele situar en la salida de la capa (tras la función de activación)

**Función activación / Función de coste**

- Clasificación binaria: activación ultima capa (sigmoide), función de perdida (binary_crossentropy)
- Clasificacion multiclase: activación ultima capa (softmax), función de perdida (categorical_crossentropy)
- Regresión con valores arbitrarios: activación ultima capa (none), función de perdida (mse)
- Regresión valores 0 a 1: activación ultima capa (sigmoide), función de perdida (mse o binary_crossentropy)

```python
# Importamos el dataset MNIST y cargamos los datos
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
n = 20
image = x_train[n,:,:]
label = y_train[n]
plt.imshow(image, cmap=plt.get_cmap('gray_r'))
plt.show()
```

```python
# Dividimos en conjunto de train, test y validación

from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split

# Cambiamos el rango, diviendo entre 255
x_train, x_te = x_train / 255.0, x_test / 255.0 

# Para convergencia --> weights y bias pequeños, y también entradas pequeñas

y_train = to_categorical(y_train, num_classes=10) #One-hot encoding para minimizar errores

print(y_train.shape)
y_te = to_categorical(y_test, num_classes=10)
x_tr, x_val, y_tr, y_val = train_test_split(x_train, y_train, test_size=0.1, random_state=42) 
print(y_tr.shape)
```

```python
# Se crea la toplogía de la red neuronal, en esta caso un MLP con una única capa de 512 neuronas
from tensorflow.keras.layers import Flatten, Dense
model = Sequential()
model.add(Flatten(input_shape=(28,28)))
model.add(Dense(512, input_shape=(28*28,), activation="relu"))
model.add(Dense(10, activation="softmax")) #Capa salida -> Mismo número de neuronas que de clases objetivo
```

Para entrenar el modelo es  necesario definir varios hiperparámetros: optimizador para functin loss,
learning rate, métrica a monitorizar --> **compile**.

```python
from tensorflow.keras.optimizers import SGD

# Se compila la arquitectura que se ha definido previamente
model.compile(loss="categorical_crossentropy", optimizer=SGD(0.005), metrics=["accuracy"])
```

**categorical_crossentropy** contrapone una clase con el resto de clases (one-vs-all)

**binario binary_crossentropy** contrapone  una clase contra la otra

```python
print(model.summary())
```

```python
# Se entrena la red neuronal
print("[INFO]: Entrenando red neuronal...")

H = model.fit(x_tr, y_tr, validation_data=(x_val, y_val), epochs=50, batch_size=128)
```

```python
import matplotlib.pyplot as plt
n = 50

%matplotlib inline

plt.style.use("ggplot")
plt.figure()
plt.plot(np.arange(0, n), H.history["loss"], label="train_loss")
plt.plot(np.arange(0, n), H.history["val_loss"], label="val_loss")
plt.plot(np.arange(0, n), H.history["accuracy"], label="train_acc")
plt.plot(np.arange(0, n), H.history["val_accuracy"], label="val_acc")
plt.title("Training Loss and Accuracy")
plt.xlabel("Epoch #")
plt.ylabel("Loss/Accuracy")
plt.legend()
```

**Como la curva de val_loss sigue a la de train_loss es un signo de que está generalizando bien, de que es un buen modelo.**

**Si la curva de val_loss se empieza a distanciar de la train_loss (normalmente a crecer cuando la otra decrece) es un signo de overfitting.**

La red pasa de optimizar los hiperparámetros para "memorizarlos".

```python
from sklearn.metrics import classification_report
# Evaluando el modelo de predicción con las imágenes de test
print("[INFO]: Evaluando red neuronal...")
predictions = model.predict(x_te, batch_size=128)
print(y_te[0])
print(predictions[0])
print(classification_report(y_te.argmax(axis=1), predictions.argmax(axis=1)))
```
